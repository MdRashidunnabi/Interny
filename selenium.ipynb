```python
#Lets import all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
```

```python
#Lets first connect to the web driver
driver=webdriver.Chrome(r"C://chromedriver.exe")
```

```python
driver.close()
```

```python
driver.get('https://www.naukri.com/')
```

```python
#Finding the element for job search bar
search_job=driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Analyst”")
search_loc=driver.find_element_by_xpath("//input[@id='qsb-location-sugg']")
search_loc.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']/button")
search_btn.click()
```

```python
#specifying the url of the webpage to be scraped
url="https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst%E2%80%9D&l=bangalore"
```

```python
#Lets open the webpage through our web driver
driver.get(url)
```

```python
job_titles=[]
company_names=[]
locations_list=[]
```

```python
#so Lets extract all the tags having the job-titles
titles_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags[0:4]
```

```python
#Now the text of the job title is inside the tags extracted above.

#so we will run a loop to iterate over the tags extracted above and extract the text inside them.

for i in titles_tags:
    titles=i.text
    job_titles.append(titles)
job_titles[0:4] 
```

```python
companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags[0:4]
```

```python
#Now the text of the companies names is inside the tags extracted above.

#so we will run a loop to iterate over the tags extracted above and extract the text inside them.

for i in companies_tags:
    companies=i.text
    company_names.append(companies)
company_names[0:4] 
```

```python
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
location_tags[0:4]
```

```python
for i in location_tags:
    locations=i.text
    locations_list.append(locations)
locations_list[0:4]
```

```python
print(len(job_titles),len(company_names),len(locations_list))
```

```python
#import pandas as pd
jobs=pd.DataFrame({})
jobs["title"]=job_titles[0:10]
jobs["company"]=company_names[0:10]
jobs["location"]=locations_list[0:10]
```

```python
jobs
```

```python
driver.close()
```
